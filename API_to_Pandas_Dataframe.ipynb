```python
import requests
import pandas as pd
from IPython.display import display
```

```python
api_key = "31cf033d"
search_query = input("Enter a movie name: ")

api_url = f"http://www.omdbapi.com/?apikey={api_key}&s={search_query}&type=movie&page=1"

response = requests.get(api_url)

if response.status_code == 200:

    data = response.json()

    if "Search" in data:

        movies = data["Search"]

        titles = []
        years = []
        directors = []
        ratings = []

        for movie in movies:
            titles.append(movie["Title"])
            years.append(movie["Year"])

            imdb_id = movie["imdbID"]
            imdb_api_url = f"http://www.omdbapi.com/?apikey={api_key}&i={imdb_id}"
            imdb_response = requests.get(imdb_api_url)

            if imdb_response.status_code == 200:
                imdb_data = imdb_response.json()
                directors.append(imdb_data.get("Director", "N/A"))
                ratings.append(imdb_data.get("imdbRating", "N/A"))
            else:
                directors.append("N/A")
                ratings.append("N/A")


        movie_data = pd.DataFrame({
            "Title": titles,
            "Year": years,
            "Director": directors,
            "Rating": ratings
        })

        top_10_movies = movie_data.sort_values(by="Rating", ascending=False).head(10)

        print("\nTop 10 Movies based on Ratings:")
        display(top_10_movies)
    else:
        print("No movie data found in the API response.")
else:
    print(f"Error: Unable to retrieve data from the OMDb API. Status Code: {response.status_code}")
```
